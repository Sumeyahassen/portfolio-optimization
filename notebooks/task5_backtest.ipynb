```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (13, 6)
```

```python
returns = pd.read_csv("../data/processed/daily_returns.csv", index_col=0, parse_dates=True)
weights_df = pd.read_csv("../data/processed/task4_portfolio_recommendation.csv", index_col=0)

print(weights_df)

# pick which column you're backtesting — e.g. "Max Sharpe" or "Min Volatility"
CHOSEN_PORTFOLIO = "Max Sharpe"   # change if you recommended Min Volatility instead

TICKERS = ["TSLA", "BND", "SPY"]
strategy_weights = weights_df.loc[TICKERS, CHOSEN_PORTFOLIO].astype(float)
print(f"\nUsing '{CHOSEN_PORTFOLIO}' weights:\n{strategy_weights}")
```

```python
BACKTEST_START = "2025-07-01"
BACKTEST_END = "2026-06-30"

backtest_returns = returns.loc[BACKTEST_START:BACKTEST_END, TICKERS]
print(f"Backtest period: {backtest_returns.index.min().date()} -> {backtest_returns.index.max().date()}")
print(f"{len(backtest_returns)} trading days")
```

```python
strategy_daily_returns = (backtest_returns * strategy_weights).sum(axis=1)
strategy_cumulative = (1 + strategy_daily_returns).cumprod()
```

```python
benchmark_weights = pd.Series({"TSLA": 0.0, "BND": 0.40, "SPY": 0.60})
benchmark_daily_returns = (backtest_returns * benchmark_weights).sum(axis=1)
benchmark_cumulative = (1 + benchmark_daily_returns).cumprod()
```

```python
fig, ax = plt.subplots(figsize=(13, 6))
ax.plot(strategy_cumulative.index, strategy_cumulative, label=f"Strategy ({CHOSEN_PORTFOLIO})")
ax.plot(benchmark_cumulative.index, benchmark_cumulative, label="Benchmark (60% SPY / 40% BND)")
ax.set_title("Cumulative Returns — Strategy vs. Benchmark")
ax.set_ylabel("Growth of $1")
ax.legend()
plt.tight_layout()
plt.savefig("../data/processed/fig_task5_backtest.png", dpi=120)
plt.show()
```
